````python
"""
演示python中的各类运算符
"""
# 算术运算符 + - * / % // **
print("1 = 1 = ", 1 + 1)
print("2 - 1 = ", 2 - 1)
print("3 * 3 = ", 3 * 3)
print("4 / 2 = ",4 / 2)
print("11 // 2 =", 11 // 2)
print("9 % 2 =", 9 % 2)
print("2 ** 2 =", 2 ** 2)
````

````python
# 赋值运算符 =
num = 1 + 2 * 3
print(num)
````

````python
# 复合赋值运算符 += -= *= /= %= **= //=
num = 1
num += 1 # num = num + 1
print("num += 1:", num)

num -= 1 
print("num -= 1:", num)

num *= 4 
print("num *= 4:", num)

num /= 2 
print("num /= 2:", num)

num = 3
num %= 2 
print("num %= 2:", num)

num **= 2 # num = num ** 2
print("num += 2:", num)

num = 9
num //= 2 # num = num // 2
print("num //= 2:", num)


````

在 Python 中，运算符用于执行各种操作，包括算术运算、比较、逻辑运算、赋值运算等。以下是 Python 中不同类型运算符的详细总结，适合初学者理解和使用。

# Python 运算符

## 1. 算术运算符
用于执行基本的数学运算。

| 运算符 | 描述          | 示例               | 结果     |
| ------ | ------------- | ------------------ | -------- |
| `+`    | 加法          | `5 + 2`            | `7`      |
| `-`    | 减法          | `5 - 2`            | `3`      |
| `*`    | 乘法          | `5 * 2`            | `10`     |
| `/`    | 除法          | `5 / 2`            | `2.5`    |
| `//`   | 取整除        | `5 // 2`           | `2`      |
| `%`    | 取模（求余）  | `5 % 2`            | `1`      |
| `**`   | 幂运算        | `5 ** 2`           | `25`     |

## 2. 比较运算符
用于比较两个值，并返回布尔值（True 或 False）。

| 运算符 | 描述          | 示例               | 结果     |
| ------ | ------------- | ------------------ | -------- |
| `==`   | 等于          | `5 == 2`           | `False`  |
| `!=`   | 不等于        | `5 != 2`           | `True`   |
| `>`    | 大于          | `5 > 2`            | `True`   |
| `<`    | 小于          | `5 < 2`            | `False`  |
| `>=`   | 大于等于      | `5 >= 2`           | `True`   |
| `<=`   | 小于等于      | `5 <= 2`           | `False`  |

## 3. 逻辑运算符
用于执行布尔逻辑操作。

| 运算符 | 描述          | 示例               | 结果     |
| ------ | ------------- | ------------------ | -------- |
| `and`  | 逻辑与        | `True and False`   | `False`  |
| `or`   | 逻辑或        | `True or False`    | `True`   |
| `not`  | 逻辑非        | `not True`         | `False`  |

## 4. 赋值运算符
用于给变量赋值。

| 运算符 | 描述          | 示例               | 结果     |
| ------ | ------------- | ------------------ | -------- |
| `=`    | 赋值          | `x = 5`            | `x` 的值为 `5` |
| `+=`   | 加法赋值      | `x += 3`           | `x` 的值加 `3` |
| `-=`   | 减法赋值      | `x -= 3`           | `x` 的值减 `3` |
| `*=`   | 乘法赋值      | `x *= 3`           | `x` 的值乘 `3` |
| `/=`   | 除法赋值      | `x /= 3`           | `x` 的值除 `3` |
| `//=`  | 取整除赋值    | `x //= 3`          | `x` 的值取整除以 `3` |
| `%=`   | 取模赋值      | `x %= 3`           | `x` 的值取模以 `3` |
| `**=`  | 幂赋值        | `x **= 3`          | `x` 的值的三次方 |

## 5. 位运算符
用于对整数进行二进制操作。

| 运算符 | 描述          | 示例               | 结果     |
| ------ | ------------- | ------------------ | -------- |
| `&`    | 按位与        | `5 & 2`            | `0`      |
| `|`    | 按位或        | `5 | 2`            | `7`      |
| `^`    | 按位异或      | `5 ^ 2`            | `7`      |
| `~`    | 按位取反      | `~5`               | `-6`     |
| `<<`   | 左移          | `5 << 1`           | `10`     |
| `>>`   | 右移          | `5 >> 1`           | `2`      |

## 6. 身份运算符
用于比较对象的存储单元（内存地址）。

| 运算符 | 描述          | 示例               | 结果     |
| ------ | ------------- | ------------------ | -------- |
| `is`   | 是            | `x is y`           | 如果 `x` 和 `y` 是同一对象，则返回 `True` |
| `is not` | 不是        | `x is not y`       | 如果 `x` 和 `y` 不是同一对象，则返回 `True` |

## 7. 成员运算符
用于测试序列中是否包含某个值。

| 运算符 | 描述          | 示例               | 结果     |
| ------ | ------------- | ------------------ | -------- |
| `in`   | 在            | `5 in [1, 2, 3, 5]`| `True`   |
| `not in` | 不在        | `4 not in [1, 2, 3, 5]` | `True` |

### 示例代码

以下是一些示例代码，展示如何使用上述运算符：

```python
# 算术运算符
a = 10
b = 3
print(a + b)  # 13
print(a - b)  # 7
print(a * b)  # 30
print(a / b)  # 3.3333...
print(a // b) # 3
print(a % b)  # 1
print(a ** b) # 1000

# 比较运算符
print(a == b) # False
print(a != b) # True
print(a > b)  # True
print(a < b)  # False
print(a >= b) # True
print(a <= b) # False

# 逻辑运算符
x = True
y = False
print(x and y) # False
print(x or y)  # True
print(not x)   # False

# 赋值运算符
c = 5
c += 2
print(c)  # 7
c -= 2
print(c)  # 5
c *= 2
print(c)  # 10
c /= 2
print(c)  # 5.0

# 位运算符
d = 5  # 0101 in binary
e = 2  # 0010 in binary
print(d & e)  # 0 (0000 in binary)
print(d | e)  # 7 (0111 in binary)
print(d ^ e)  # 7 (0111 in binary)
print(~d)     # -6
print(d << 1) # 10 (1010 in binary)
print(d >> 1) # 2 (0010 in binary)

# 身份运算符
f = [1, 2, 3]
g = f
h = [1, 2, 3]
print(f is g)     # True
print(f is h)     # False
print(f is not h) # True

# 成员运算符
i = [1, 2, 3, 4, 5]
print(3 in i)     # True
print(6 not in i) # True
```
